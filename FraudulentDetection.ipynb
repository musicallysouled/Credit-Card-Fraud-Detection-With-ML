# Credit Card Fraud Detection

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv", sep=",")
```

```python
data.head()
```

```python
data.tail()
```

```python
data.info()
```

This shows information is related to the dataset. No missing values found and the datatype is float.

```python
#distribution of legitimate and fraudulent transactions
data['Class'].value_counts()
```

The above shows that the dataset is highly unbalanced.

```python
#for further analysis
legit = data[data.Class == 0]
fraud = data[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical view
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#comparing the values
data.groupby('Class').mean()
```

Building sample dataset with uniform distribution

```python
sample_legit=legit.sample(n=56)
#same as number of fraudulent transactions. This will provide us random data points
```

```python
#concatenation of dataframes
new_data = pd.concat([sample_legit, fraud], axis = 0)
```

```python
#axis=0? : data will be entered row wise
# if axis =1 : data will be entered column wise
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

Therefore, the data is uniformly distributed.

```python
new_data.groupby('Class').mean()
```

# Splitting the data(features and targets)

```python
X= new_data.drop(columns= 'Class', axis=0)
Y= new_data['Class']
```

```python
print(X)
```

```python
print(Y)
```

# Train & Test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, stratify = Y, random_state= 2)
#test size is 20%, 80% training data; random state splits data in two as mentioned
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training - Logistic Regression

```python
model = LogisticRegression()

# Training the data
model.fit(X_train, Y_train)
```

# Evaluating model - Accuracy score

```python
X_train_predict = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predict, Y_train)
```

```python
print('Accuracy score of training data is: ',training_data_accuracy)
```

```python
X_test_predict = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_predict, Y_test)
print('Accuracy score of test data is: ',testing_data_accuracy)
```
